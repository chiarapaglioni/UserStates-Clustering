```python
#groups
users = filtered_df['User ID'].unique()
user_groups = filtered_df[filtered_df['User ID'].isin(users)].groupby('User ID')

# Loop through each user's group
for user_id, group in user_groups:
    
    #STEPS AND SLEEP AND CALLORIES BURN FOR EACH USER

    # Create a new figure
    plt.figure(figsize=(10, 6))
    
    # Plot Steps
    plt.plot(group['Date/Time'], group['Steps'], marker='s', linestyle='--', color='green', label='Steps')

    # Plot Sleep
    plt.plot(group['Date/Time'], group['Sleep'], marker='^', linestyle='-.', color='orange', label='Sleep (?)')

    # Callories Burn
    plt.plot(group['Date/Time'], group['Callories Burn'], marker='^', linestyle='-.', color='blue', label='Callories Burn')

    # Plot styling
    plt.title(f'Steps, Sleep & Callories Burn for User {user_id}')
    plt.xlabel('Timeframe')
    plt.ylabel('Values')
    plt.xticks(rotation=45)
    plt.legend()
    plt.grid(True)

    # Show the plot
    plt.show()

    #HRV AND SLEEP HEART RATE PER USER

    # Create a new figure
    plt.figure(figsize=(10, 6))
    
    # Plot HRV
    plt.plot(group['Date/Time'], group['HRV'], marker='s', linestyle='--', color='green', label='HRV')

    # Plot Sleep Heart Rate
    plt.plot(group['Date/Time'], group['Sleep Heart Rate'], marker='^', linestyle='-.', color='orange', label='Sleep Heart Rate')

    # Plot styling
    plt.title(f'HRV & Sleep Heart Rate for User {user_id}')
    plt.xlabel('Timeframe')
    plt.ylabel('Values')
    plt.xticks(rotation=45)
    plt.legend()
    plt.grid(True)

    # Show the plot
    plt.show()

    # PERFORMANCE FOR EACH USER

    # Create a new figure
    plt.figure(figsize=(10, 6))
    
    # Plot Performance
    plt.plot(group['Date/Time'], group['Performance'], marker='o', linestyle='-', color='blue', label='Performance')

    # Plot styling
    plt.title(f'Performance for User {user_id}')
    plt.xlabel('Timeframe')
    plt.ylabel('Values')
    plt.xticks(rotation=45)
    plt.legend()
    plt.grid(True)

    # Show the plot
    plt.show()
```
