```python
import matplotlib.pyplot as plt

# Data for the pie chart
emotion_targets = ['Happy', 'Determined', 'Excited', 'Loving', 'Motivated', 'Powerful',
                   'Helpful', 'Lively', 'Peaceful', 'Proud', 'Optimistic', 'Valued',
                   'Content', 'Hopeful', 'Confident', 'Relieved', 'Annoyed', 'Nervous',
                   'Overwhelmed', 'Irritated', 'Anxious', 'Afraid', 'Angry', 'Worried',
                   'Bored', 'Lonely', 'Weary', 'Disappointed', 'Confused', 'Embarrassed', 'Sad']

wearable_data = ['Steps', 'Sleep', 'HRV', 'Sleep Heart Rate',
                 'Glucose', 'Callories Burn', 'Calories Intake']

general_data = ['User ID', 'Phone', 'Gender', 'Age', 'Weight', 'Height', 'Tiredness', 'Calm', 'Nutrition',
                'Hydration', 'Performance', 'Concentrate']

text_data = ['Actions', 'With', 'Reflections']

# Define category sizes
emotion_categories = {
    'Emotion Targets': len(emotion_targets),
    'Wearable Data': len(wearable_data),
    'User Data': len(general_data), 
    'Text Data': len(text_data), 
}

# Create pie chart
labels = list(emotion_categories.keys())
sizes = list(emotion_categories.values())

# Set up the figure size
plt.figure(figsize=(8, 6))

# Create the pie chart
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140)

# Equal aspect ratio ensures that pie is drawn as a circle
plt.axis('equal')

# Add a legend to the side of the pie chart
plt.legend(labels, title="Categories", bbox_to_anchor=(1.05, 0.5), loc='center left')

# Save the plot
plot_path = "plots/pie-chart.png"
plt.savefig(plot_path, bbox_inches='tight')

# Show the plot
plt.show()
```
